# Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X = Data.drop('Power_Generated',axis=1)
y= Data['Power_Generated']
```

```python
X = sc.fit_transform(X)
```

```python
y = sc.fit_transform(np.array(y).reshape(-1,1))
```

```python
X
```

```python
y
```

# Splitting the Data set into training and testing data set

```python
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Building the LSTM Model

```python
import tensorflow as tf 
from tensorflow import keras    
from keras.models import Sequential
from keras.layers import Dense  
from keras.layers import LSTM 
```

```python
Model = Sequential()
Model.add(LSTM(300,activation='relu',return_sequences=True,input_shape=(11,1)))
Model.add(LSTM(300,activation='relu'))
Model.add(Dense(1))
Model.compile(optimizer='adam', loss='mae')
history = Model.fit(X_train,y_train,epochs=10)
```

```python
pd.DataFrame(history.history).plot()
```

```python
Model.evaluate(X_test,y_test)
```
